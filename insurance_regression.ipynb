```python
# Import all the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the insurance.csv file
df = pd.read_csv("insurance.csv")
```

```python
# Check first few observations
df.head()
```

```python
# Check the shape of the dataframe
df.shape
```


Now, plot a scatter plot with age on the X-axis and charges on the Y-axis

```python
# creating a scatterplot and display the plot
plt.scatter(x = df.age, y = df.charges)
#sns.scatterplot(data = df, x = df.age, y = df.charges)
plt.title("Age vs Insurance Charges")
plt.xlabel("Age")
plt.ylabel("Insurance Charges")
plt.show()
```

As you can see above, it clearly indicates that insurance cost increases as age increases

Now extract the required columns for linear regression model

```python
# Select the column to build linear regression model
X = df.iloc[:,[0]].values
y = df.iloc[:,[6]].values
```

```python
# Create a Linear regression model
insurance_model = LinearRegression()
```

```python
# Fit the model
insurance_model.fit(X, y)
```

```python
# Making predictions
y_pred = insurance_model.predict(X)
```

```python
# Plot another scatter plot with the best-fit line
plt.scatter(X,y,color = 'b')
plt.plot(X,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.show()
```
