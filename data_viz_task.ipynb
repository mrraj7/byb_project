```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
cars_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

```python
# Select the 'Audi', 'Hyundai', 'Suzuki', 'Toyota' manufacturers to create cars dataset
cars = cars_data[cars_data['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]
```

```python
# Which manufacturer in cars dataset has the highest revs per mile
cars.groupby('Manufacturer')['Rev.per.mile'].max()
```

### 1. Boxplot

```python
# Boxplot for revs per mile that shows Which manufacturer in cars dataset has the highest revs per mile
sns.boxplot(data=cars, x='Manufacturer', y='Rev.per.mile')
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
From the above boxplot graph, it showsToyota has the highest revs per mile

### 2. Histogram

```python
# Histogram of MPG
df = cars[['MPG.city','MPG.highway']]
sns.histplot(data = df, multiple = "dodge")
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Highway as per above histogram

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(data = cars, x = 'Wheelbase', y = 'Turn.circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
When the wheelbase get larger, turn-circle increases.

### 4. Barplot

```python
# Barplot for mean horsepower
#sns.barplot(data=cars, x = 'Type', y = cars.groupby('Type')['Horsepower'].mean())
cars_hp_mean = cars.groupby('Type')['Horsepower'].mean()
ax = sns.barplot(data = cars_hp_mean)
ax.set(xlabel = 'Car Type', ylabel='Mean of Horsepower', title = 'Mean of Horsepower vs Car Type')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
No
